```python
import requests
from urllib import parse
from pprint import pprint
import pandas as pd
import config
import numpy as np

input_data = 'data/cleaned_opendoar_data.csv'

df_all_data = pd.read_csv(input_data, header=0, index_col='system_metadata.id')
display(df_all_data)

 
```
